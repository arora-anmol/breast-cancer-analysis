```python
# Create correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.90
to_drop = [column for column in upper.columns if any(upper[column] > 0.90)]
to_drop
```

```python
df.drop(columns = to_drop, inplace=True)
```

```python
labels = df['diagnosis']
df.drop(columns = ['diagnosis'], inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, labels)
```

```python
df_b.shape
```

```python
df_m.shape
```

```python
clf_naive_bayes = GaussianNB()
```

```python
clf_naive_bayes.fit(x_train, y_train)
```

```python
y_pred = clf_naive_bayes.predict(x_test)
```

```python
accuracy_naive_bayes = accuracy_score(y_test, y_pred)
accuracy_naive_bayes
```

```python
clf_svm = svm.SVC(C = 1, kernel='rbf', gamma='auto')
```

```python
clf_svm
```

```python
clf_svm.fit(x_train, y_train)
```

```python
y_pred = clf_svm.predict(x_test)
accuracy_rbf_svm = accuracy_score(y_test, y_pred)
```

```python
accuracy_rbf_svm
```

```python
clf_svm_linear = svm.SVC(kernel='linear', C = 2000)
```

```python
clf_svm_linear.fit(x_train, y_train)
```

```python
y_pred = clf_svm_linear.predict(x_test)
accuracy_linear_svm = accuracy_score(y_test, y_pred)
accuracy_linear_svm
```

```python
p1 = plt.bar( ['linear_svm', 'rbf_svm', 'naive_bayes'], [accuracy_linear_svm, accuracy_rbf_svm, accuracy_naive_bayes])
plt.ylabel('accuracy')
plt.xlabel('classifier')
plt.show()
```

```python
y_pred_train = clf_svm_linear.predict(x_train)
```
